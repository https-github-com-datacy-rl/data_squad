# Visualisation with Matplotlib and Seaborn

Matplotlib and Seaborn are two of the most popular Python libraries for data visualization. They provide a wide array of functionalities to create various visual representations of data, which can be instrumental in understanding complex data sets and conveying insights.

## Matplotlib Basics
Matplotlib is a low-level graph plotting library in Python that serves as a visualization utility. It's highly customizable but can also be complex to use for intricate plots.

### Installation
If you haven't installed Matplotlib, you can do so using pip:

```python
# pip install matplotlib
```

### Basic Plot
Creating a simple line plot in Matplotlib:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Scatter Plot
Creating a scatter plot:

```python
plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## Seaborn Basics
Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics.

### Installation
To install Seaborn, use pip:

```python
# pip install seaborn
```

### Basic Plot
Creating a simple Seaborn plot:

```python
import seaborn as sns

# Load example dataset
iris = sns.load_dataset("iris")

# Create a simple Seaborn plot
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.show()
```

## Combining Matplotlib and Seaborn
Matplotlib can be used to customize Seaborn plots further:

```python
# Using Matplotlib to set the figure size and titles for a Seaborn plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.title("Iris Sepal Dimensions")
plt.show()
```

## More Advanced Plots
### Histograms in Matplotlib

```python
plt.hist(iris["sepal_length"], bins=20)
plt.title("Sepal Length Histogram")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

### Boxplot in Seaborn

```python
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Sepal Length by Species")
plt.show()
```

### Pairplot in Seaborn
This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame. It's a great way to get a quick overview of the dataset.

```python
sns.pairplot(iris, hue="species")
plt.show()
```

## Conclusion

- Seaborn simplifies the process of creating sophisticated statistical plots.
- Both libraries work well together, allowing you to leverage the simplicity of Seaborn and the power of Matplotlib for fine-tuning.
These libraries are essential tools in the Python data visualization landscape and are widely used for academic and professional purposes. They can greatly assist in exploring and understanding data, as well as communicating data-driven findings effectively.
